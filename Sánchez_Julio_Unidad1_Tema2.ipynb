## Ejercicio 1: Sistema de Votación

Programa que simula un proceso de votación sencillo. El usuario ingresa votos de forma continua:
- `1` → Candidato A
- `2` → Candidato B  
- `3` → Candidato C
- `0` → Finalizar votación

**Estructuras utilizadas:** `while` (repetición), `if-elif-else` (decisión)

```python
# ============================================================
# EJERCICIO 1: SISTEMA DE VOTACIÓN
# ============================================================

# Contadores de votos para cada candidato
votos_a = 0
votos_b = 0
votos_c = 0

print("=" * 40)
print("      SISTEMA DE VOTACIÓN")
print("=" * 40)
print("1 = Candidato A")
print("2 = Candidato B")
print("3 = Candidato C")
print("0 = Finalizar votación")
print("=" * 40)

# Ciclo while: continúa recibiendo votos hasta que se ingrese 0
while True:
    try:
        voto = int(input("\nIngrese su voto: "))
    except ValueError:
        print("Error: Ingrese un número válido.")
        continue

    # Estructura de decisión para registrar o rechazar el voto
    if voto == 1:
        votos_a += 1          # Incrementa contador del Candidato A
        print("✓ Voto registrado para Candidato A.")
    elif voto == 2:
        votos_b += 1          # Incrementa contador del Candidato B
        print("✓ Voto registrado para Candidato B.")
    elif voto == 3:
        votos_c += 1          # Incrementa contador del Candidato C
        print("✓ Voto registrado para Candidato C.")
    elif voto == 0:
        print("Finalizando votación...")  # Señal de parada
        break                 # Sale del ciclo while
    else:
        # Voto inválido: no se suma a ningún candidato
        print("✗ Voto inválido. Solo se permiten opciones 0, 1, 2 o 3.")

# Mostrar resultados finales
print("\n" + "=" * 40)
print("        RESULTADOS FINALES")
print("=" * 40)
print(f"  Candidato A: {votos_a} votos")
print(f"  Candidato B: {votos_b} votos")
print(f"  Candidato C: {votos_c} votos")
print("=" * 40)
print("La votación ha finalizado.")
```

## Ejercicio 2: Sistema de Métodos de Pago

Programa que simula un sistema de compras con distintos métodos de pago:
- `1` → Tarjeta de crédito (+5% comisión)
- `2` → Tarjeta de débito (0% comisión)
- `3` → Efectivo (−2% descuento)
- `4` → Salir

**Estructuras utilizadas:** `while` (repetición), `if-elif-else` (decisión), validación de monto

```python
# ============================================================
# EJERCICIO 2: SISTEMA DE MÉTODOS DE PAGO
# ============================================================

print("=" * 45)
print("      SISTEMA DE MÉTODOS DE PAGO")
print("=" * 45)

# Ciclo principal: se repite hasta que el usuario elija Salir (opción 4)
while True:
    print("\n--- MENÚ DE PAGO ---")
    print("1. Pagar con tarjeta de crédito  (+5% comisión)")
    print("2. Pagar con tarjeta de débito   (sin comisión)")
    print("3. Pagar en efectivo             (-2% descuento)")
    print("4. Salir del sistema")

    try:
        opcion = int(input("Seleccione una opción: "))
    except ValueError:
        print("Error: Ingrese un número válido.")
        continue

    if opcion == 4:
        # El usuario elige salir del sistema
        print("\nGracias por usar el sistema. ¡Hasta luego!")
        break

    elif opcion in (1, 2, 3):
        # Solicitar y validar el monto de la compra
        try:
            monto = float(input("Ingrese el monto de la compra: $"))
        except ValueError:
            print("Error: Ingrese un valor numérico.")
            continue

        # Validar que el monto sea mayor a 0
        if monto <= 0:
            print("Error: El valor ingresado no es válido. Debe ser mayor a 0.")
            continue

        # Calcular monto final según el método elegido
        if opcion == 1:
            comision = monto * 0.05          # 5% de comisión por crédito
            monto_final = monto + comision
            metodo = "Tarjeta de Crédito"
            detalle = f"Comisión (5%): +${comision:.2f}"
        elif opcion == 2:
            monto_final = monto              # Sin comisión para débito
            metodo = "Tarjeta de Débito"
            detalle = "Comisión (0%): $0.00"
        else:
            descuento = monto * 0.02         # 2% de descuento por efectivo
            monto_final = monto - descuento
            metodo = "Efectivo"
            detalle = f"Descuento (2%): -${descuento:.2f}"

        # Mostrar resumen de la operación
        print("\n" + "-" * 35)
        print(f"  Método de pago:  {metodo}")
        print(f"  Monto original:  ${monto:.2f}")
        print(f"  {detalle}")
        print(f"  Monto final:     ${monto_final:.2f}")
        print("-" * 35)

    else:
        # Opción fuera del rango permitido
        print("Opción inválida. Seleccione entre 1 y 4.")
```

## Ejercicio 3: Calculadora Avanzada

Programa que simula una calculadora con las siguientes operaciones:
1. Sumar | 2. Restar | 3. Multiplicar | 4. Dividir | 5. Potencia (A^B) | 6. Raíz cuadrada (√A) | 7. Salir

**Estructuras utilizadas:** `while` (repetición principal), `while` anidado (validación), `if-elif-else` (decisión)

```python
# ============================================================
# EJERCICIO 3: CALCULADORA AVANZADA
# ============================================================

import math  # Módulo para pow() y sqrt()

print("=" * 40)
print("        CALCULADORA AVANZADA")
print("=" * 40)

# Ciclo principal: muestra el menú hasta que el usuario elija Salir
while True:
    print("\n--- MENÚ DE OPERACIONES ---")
    print("1. Sumar")
    print("2. Restar")
    print("3. Multiplicar")
    print("4. Dividir")
    print("5. Potencia (A^B)")
    print("6. Raíz cuadrada (√A)")
    print("7. Salir")

    try:
        opcion = int(input("Seleccione una opción: "))
    except ValueError:
        print("Error: Ingrese un número válido.")
        continue

    if opcion == 7:
        print("Saliendo de la calculadora. ¡Hasta luego!")
        break

    elif opcion == 1:
        # Operación: SUMA
        a = float(input("Ingrese A: "))
        b = float(input("Ingrese B: "))
        resultado = a + b
        print(f"Operación: SUMA → {a} + {b} = {resultado}")

    elif opcion == 2:
        # Operación: RESTA
        a = float(input("Ingrese A: "))
        b = float(input("Ingrese B: "))
        resultado = a - b
        print(f"Operación: RESTA → {a} - {b} = {resultado}")

    elif opcion == 3:
        # Operación: MULTIPLICACIÓN
        a = float(input("Ingrese A: "))
        b = float(input("Ingrese B: "))
        resultado = a * b
        print(f"Operación: MULTIPLICACIÓN → {a} × {b} = {resultado}")

    elif opcion == 4:
        # Operación: DIVISIÓN — valida que el divisor no sea cero
        a = float(input("Ingrese A (dividendo): "))
        while True:
            b = float(input("Ingrese B (divisor, ≠ 0): "))
            if b != 0:
                break          # Divisor válido, sale del ciclo de validación
            print("Error: No se puede dividir entre cero. Intente de nuevo.")
        resultado = a / b
        print(f"Operación: DIVISIÓN → {a} / {b} = {resultado:.4f}")

    elif opcion == 5:
        # Operación: POTENCIA — A debe ser positivo
        while True:
            a = float(input("Ingrese A (debe ser positivo): "))
            if a > 0:
                break          # Valor válido, sale del ciclo de validación
            print("Error: A debe ser un valor positivo. Intente de nuevo.")
        b = float(input("Ingrese B (exponente): "))
        resultado = math.pow(a, b)   # Cálculo de la potencia
        print(f"Operación: POTENCIA → {a}^{b} = {resultado}")

    elif opcion == 6:
        # Operación: RAÍZ CUADRADA — A debe ser positivo
        while True:
            a = float(input("Ingrese A (debe ser positivo): "))
            if a > 0:
                break          # Valor válido, sale del ciclo de validación
            print("Error: A debe ser un valor positivo. Intente de nuevo.")
        resultado = math.sqrt(a)     # Cálculo de la raíz cuadrada
        print(f"Operación: RAÍZ CUADRADA → √{a} = {resultado:.4f}")

    else:
        # Opción fuera del rango válido: muestra error y vuelve al menú
        print("Error: Opción inválida. Seleccione entre 1 y 7.")
        continue

    # Preguntar si desea continuar (solo después de una operación válida)
    if opcion in (1, 2, 3, 4, 5, 6):
        continuar = input("\n¿Desea realizar otra operación? (s/n): ").strip().lower()
        if continuar != 's':
            print("Saliendo de la calculadora. ¡Hasta luego!")
            break
```

## Ejercicio 4: Control de Acceso — Sistema de Seguridad

Programa que simula un sistema de seguridad para un edificio:
- Máximo **3 intentos** de autenticación
- Si falla → `"Acceso bloqueado"`
- Si entra → menú con: ver estado, cambiar contraseña, salir

**Estructuras utilizadas:** `while` + contador de intentos, `if-elif-else` anidado, `while` para confirmación de contraseña

> 🔑 **Credenciales:** usuario: `admin` / contraseña: `1234`

```python
# ============================================================
# EJERCICIO 4: CONTROL DE ACCESO — SISTEMA DE SEGURIDAD
# ============================================================

# Credenciales predefinidas del sistema
usuario_correcto = "admin"
contrasenia_sistema = "1234"

intentos = 0          # Contador de intentos fallidos
acceso_ok = False     # Bandera de acceso exitoso
MAX_INTENTOS = 3      # Número máximo de intentos permitidos

print("=" * 45)
print("    SISTEMA DE CONTROL DE ACCESO")
print("=" * 45)
print("Por favor ingrese sus credenciales.")

# Ciclo de autenticación: permite hasta MAX_INTENTOS intentos
while intentos < MAX_INTENTOS and not acceso_ok:
    usuario = input("\nUsuario: ").strip()
    contrasenia = input("Contraseña: ").strip()

    intentos += 1  # Contar este intento

    if usuario == usuario_correcto and contrasenia == contrasenia_sistema:
        acceso_ok = True   # Credenciales correctas
    else:
        # Credenciales incorrectas: mostrar intentos restantes
        restantes = MAX_INTENTOS - intentos
        print(f"✗ Usuario o contraseña incorrectos.")
        if restantes > 0:
            print(f"  Intentos restantes: {restantes}")

# Verificar resultado de la autenticación
if not acceso_ok:
    # Se superó el número máximo de intentos
    print("\n" + "!" * 45)
    print("           ACCESO BLOQUEADO")
    print("!" * 45)
else:
    # Acceso concedido: mostrar bienvenida
    print(f"\n✓ Bienvenido, {usuario}! Acceso concedido al edificio.")

    # Ciclo del menú principal (se repite hasta elegir Salir)
    while True:
        print("\n--- MENÚ DE OPCIONES ---")
        print("1. Ver estado de acceso")
        print("2. Cambiar contraseña")
        print("3. Salir")

        try:
            opcion_menu = int(input("Seleccione una opción: "))
        except ValueError:
            print("Error: Ingrese un número válido.")
            continue

        if opcion_menu == 1:
            # Mostrar estado de acceso del usuario autenticado
            print(f"\nEstado: El usuario '{usuario}' tiene acceso autorizado al edificio.")

        elif opcion_menu == 2:
            # Proceso de cambio de contraseña
            nueva = input("Ingrese la nueva contraseña: ").strip()

            # Ciclo de confirmación: repite hasta que ambas contraseñas coincidan
            while True:
                confirmacion = input("Confirme la nueva contraseña: ").strip()
                if nueva == confirmacion:
                    contrasenia_sistema = nueva   # Actualizar contraseña en el sistema
                    print("✓ Contraseña actualizada exitosamente.")
                    break
                else:
                    # No coinciden: permitir reintentar la confirmación
                    print("✗ Las contraseñas no coinciden. Intente de nuevo.")

        elif opcion_menu == 3:
            # Cerrar sesión y finalizar el programa
            print(f"\nSesión cerrada. ¡Hasta luego, {usuario}!")
            break

        else:
            print("Opción inválida. Seleccione 1, 2 o 3.")
```
