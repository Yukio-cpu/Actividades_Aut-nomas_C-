```python
promedio = float(input("Introduce tu promedio: "))
nvl = str(input("Introduce tu nivel: (Universitario = 1 ,Bachillerato = 2, Básico = 3) "))
if promedio >= 9 and nvl == "1":
    print("Felicidades, obtuviste una beca!!!")
elif 8.0 <= promedio <= 8.9 and nvl == "2":
    print("Felicidades, obtuviste una beca!!!")
else:
    print("Lo siento, no obtuviste una beca.")
print("Tu promedio es: ", promedio)
print("Tu nivel es: ", nvl)
```

```python
costo = float(input("Introduce el costo de la carrera: "))
costo_final = costo
tipo = str(input("Introduce el tipo de estudiante: (Regular = 1, Becado = 2, Extraordinario = 3) "))
if costo > 1500 :
    print("Tu costo base es mayor a 1500, tienes un descuento del 10%")
    costo_final = costo - (costo * 0.1)
if tipo == "1":
    print("Eres un estudiante regular, tu recargo es del 15%")
    costo_final = costo_final + (costo_final * 0.15)
elif tipo == "2":
    print("Eres un estudiante becado, tu recargo es del 5%")
    costo_final = costo_final + (costo_final * 0.05)
elif tipo == "3":
    print("Eres un estudiante extraordinario, tu recargo es del 25%")
    costo_final = costo_final + (costo_final * 0.25)
print("Tu costo base es: ", costo)
print("Tu costo final es: ", costo_final)
```

```python
vel = float(input("Introduce tu velocidad en km/h: "))
cont = "null"
while vel < 0 or vel > 300:
    print("Velocidad inválida. Introduce una velocidad entre 0 y 300 km/h.")
    vel = float(input("Introduce tu velocidad en km/h: "))
conversion = vel / 3.6 
if vel < 40 :
    cont = "Lenta"
elif 40 <= vel <= 80 :
    cont = "Moderada"
elif 80 < vel <= 120 :
    cont = "Rápida"
else:
    cont = "Muy rápida"
print("Tu velocidad en km/h es: ", vel)
print("Tu velocidad en m/s es: ", conversion)
print("Tu velocidad es : ", cont)
```

```python
import random
numero_secreto = random.randint(0, 10)
intentos = 0
max_intentos = 5
while intentos < max_intentos:
    numero = int(input("Adivina el número secreto: "))
    intentos += 1
    if numero == numero_secreto:
        print(f"¡Correcto! Lo lograste en {intentos} intentos")
        break
    else:
        print("Demasiado alto" if numero > numero_secreto else "Demasiado bajo")
if intentos >= max_intentos:
    print(f"Lo siento, se agotaron los intentos. El número era {numero_secreto}")
```
